📌Prueba la hipótesis:

"La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos".

Hipótesis nula (H₀):
La duración promedio de los viajes desde el Loop hacia el Aeropuerto O'Hare no cambia en los sábados lluviosos respecto a los sábados sin lluvia.
→ μ_lluvioso = μ_seco

Hipótesis alternativa (H₁):
La duración promedio sí cambia los sábados lluviosos.
→ μ_lluvioso ≠ μ_seco

```python
print(df3['weather_conditions'].unique())
```

```python
def clim(df):
    array_good = []
    array_bad = []
    
    for index, row in df.iterrows():
        if row['weather_conditions'] == 'Good':
            array_good.append(row['duration_seconds'])
        elif row['weather_conditions'] == 'Bad':
            array_bad.append(row['duration_seconds'])
    
    return array_good, array_bad


def adorno(caract='*', nume=40, letrero=None):
    if not letrero:
        print(caract * (nume * 2))
    else:
        print(caract * nume + letrero + caract * nume)


#para llamarlas a modo de ejemplo
#adorno()  # Línea decorativa simple
#adorno('=', 30, ' df_games ')  # Decoración con letrero centrado
```

```python
good_duracion, bad_duracion = clim(df3)

#imprimi las listas para mostra y tratan de entender un poco mas

adorno('*', 50) 
adorno('*', 40,'clima bueno') 
print()
print(good_duracion)
print()
adorno('*', 40,'clima malo') 
print()
print(bad_duracion)
print()
adorno('*', 50) 
#print(df3)
```
